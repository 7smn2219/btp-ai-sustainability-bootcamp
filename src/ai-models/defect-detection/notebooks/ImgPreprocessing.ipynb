## Preprocessing

```python
import pandas as pd
import cv2
```

```python
import glob
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
files=glob.glob('../data/Images/NG/*')
```

```python
path=files[16]
im=cv2.imread(path,0)

clahe = cv2.createCLAHE(clipLimit=5.0, tileGridSize=(5,5))
cl1 = clahe.apply(im)

kernel = np.ones((3,3),np.uint8)
dilation = cv2.dilate(cl1,kernel,iterations = 1)

kernel = np.ones((5,5),np.uint8)
dilation5 = cv2.dilate(cl1,kernel,iterations = 1)


fig,ax=plt.subplots(1,4, figsize=(15,60))
ax[0].imshow(im)
ax[1].imshow(cl1)
ax[2].imshow(dilation)
ax[3].imshow(dilation5)
```
